```python
###Got this from Guillaume Cantin's website, and have been messing around with it
### to better understand our good friend Henon and his map


%matplotlib inline
from matplotlib import pyplot as plt
from random import randint, random
import numpy as np

### parameters, these are fun to play with, just make sure you iterate enough times
### to get a good grasp of whats going on
a = 1.4
b = 0.3

### Henon map, here is the good good math equations being used, playing with this is also fun

def henon(X):
    x, y = X
    xnew = a-x*x+b*y
    ynew = x
    return xnew, ynew

X0 = [(.5-b)/2, (.5-b)/2]

X, Y = [], []
### messing with the range/iteration, you can see how the Henon map really shows
### the shape and attraction
for i in range(10000):
    xnew, ynew = henon(X0)
    X, Y = X + [xnew], Y + [ynew]
    X0 = [xnew, ynew]

### Here it is plotted out
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(X, Y, color='orange', s=.2)
plt.show()
```
