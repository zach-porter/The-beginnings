```python
%matplotlib inline ### Run this code, it allows for the visuals/graphs to pop up
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(c,x): return c*x*(1-x)
n = 100 ###changing n can adjust how many times we iterate the function
c = 3.3 ### play with this as well, mainly going between numbers from 2-4
x0 = .00000000001 ### I start real small here, but you can change up how small want this
x1 = x0
cobweb_xs = [x1]
cobweb_ys = [x1]
```

```python
for i in range(n):
    x2 = f(c,x1)
    cobweb_xs.append(x1)
    cobweb_ys.append(x2)
    x1 = x2
    cobweb_xs.append(x1)
    cobweb_ys.append(x2)
graph_xs = np.linspace(-1,1,100)
graph_ys = f(c, graph_xs)
start = x1
end = x2
```

```python
for i in range(1,len(cobweb_xs)): ###This is mainly about the graph, and how to adjust it
    plt.plot([cobweb_xs[i-1],cobweb_xs[i]],[cobweb_ys[i-1],cobweb_ys[i]],'green', alpha=0.5, linewidth=2)
plt.plot([0,1],[0,1],'--', linewidth=3)
plt.plot(graph_xs,graph_ys, 'blue', linewidth=5)
plt.plot(x0,x0, 'go', zorder=5)
plt.plot(x2,x2, 'ro', zorder=5)
plt.title("Cobweb Plot")

ax = plt.gca()
ax.set_aspect(1)
ax.set_ylim(0,1)
ax.set_xticks([0,.25,.5,.75,1])
ax.set_yticks([0,.25,.5,.75,1])
ax.set_xlim(0,1)

fig = ax.figure
fig.set_figwidth(8)
fig.set_figheight(8)
```
